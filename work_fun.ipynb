```python
import numpy as np
import matplotlib.pyplot as plt

# Carregar dados do arquivo txt
data = np.loadtxt('avg.dat')

# Separar colunas
coluna2 = data[:, 0]  # Coluna 1 como eixo y
coluna1 = data[:, 1]  # Coluna 2 como eixo x

# Criar o gráfico
plt.figure(figsize=(8, 6))
plt.plot(coluna2, coluna1, marker='o', linestyle='-', color='b')

# Adicionar título e rótulos aos eixos
plt.title('Coordenada da Superfície (111) em Z vs Valor Médio da Energia Potencial')
plt.xlabel('Coordenadas em Z')
plt.ylabel('Valor médio de V')

# Mostrar o gráfico
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constante de conversão de Ry para eV
rydberg_to_ev = 13.6056980659

# Carregar dados do arquivo txt
data = np.loadtxt('avg.dat')

# Separar colunas
coluna2 = data[:, 0]  # Coluna 2 continua sendo eixo x
coluna1 = data[:, 1]  # Coluna 1 será convertida de Ry para eV

# Converter a coluna 1 de Ry para eV
coluna1_ev = coluna1 * rydberg_to_ev

# Criar o gráfico com as unidades convertidas
plt.figure(figsize=(8, 6))
plt.plot(coluna2, coluna1_ev, marker='o', linestyle='-', color='b')

# Adicionar título e rótulos aos eixos
plt.title('Coordenada da Superfície (111) em Z vs Valor Médio da Energia Potencial')
plt.xlabel('Coordenadas em Z (a.u)')
plt.ylabel('Valor médio de V (eV)')

# Mostrar o gráfico
plt.grid(True)
plt.savefig('grafico_potencial.png', dpi=300)
plt.show()
```
